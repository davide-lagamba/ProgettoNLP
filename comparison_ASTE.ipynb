```python
#fonte:
#https://pyabsa.readthedocs.io/en/latest/6_tutorials/Aspect_Sentiment_Triplet_Extraction.html
from pyabsa import available_checkpoints
from pyabsa import AspectSentimentTripletExtraction as ASTE

checkpoints = available_checkpoints(show_ckpts=True)
triplet_extractor = ASTE.AspectSentimentTripletExtractor(checkpoint="multilingual")
```

```python
#Upload dataset "test_final.txt" in "/content"
import re
phrases = []

test_file = open("test_final.txt", "r")
content = test_file.read()
content_new = re.sub('#### #### ####', '####', content, flags=re.M)
output = open("test_final_modified.txt", "w")
output.write(content_new)

test_file_modified = open("test_final_modified.txt", "r")
lines = test_file_modified.readlines()
for i in lines:
  phrases.append(i)
```

```python
pred = 0
test = 0
correct_triples = 0
errors = 0

for i in phrases:
  print(i)
```

```python
from types import NoneType

#fonte:
#https://github.com/chiayewken/Span-ASTE  file "aste/wrapper.py" metodo "score"

for i in range(len(phrases)):
    triple = triplet_extractor.predict(phrases[i])
    print(triple)
    if(type(triple.get("Triplets")) is NoneType or type(triple.get("True Triplets")) is NoneType):
      errors += 1
      continue
    pred += len(triple.get("Triplets"))
    test += len(triple.get("True Triplets"))
    for j in triple.get("Triplets"):
      for z in triple.get("True Triplets"):
        if j == z:
          correct_triples += 1

precision = correct_triples/pred
recall = correct_triples/test
f1 = (2*precision*recall)/(precision + recall)
```

```python
print("precision: "+str(precision))
print("recall: " +str(recall))
print("f1: " +str(f1))
print("errors: "+str(errors))
```
